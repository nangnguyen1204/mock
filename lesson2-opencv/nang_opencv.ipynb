```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
img = cv2.imread("index.jpg", cv2.IMREAD_COLOR)
img1 = cv2.imread("output.png")

rgb_img = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

display_img(img, None)
display_img(img1, None)
display_img(rgb_img, None)
```

```python
img.shape, rgb_img.shape
```

```python
# to convert the image in grayscale 
img2 = cv2.imread('output.png', cv2.IMREAD_GRAYSCALE)

ret, thresh21 = cv2.threshold(img2, 120, 255, cv2.THRESH_BINARY)


display_img(img2, None)
display_img(thresh21, None)
```

```python
hsv_img = cv2.cvtColor(rgb_img, cv2.COLOR_RGB2HSV)

display_img(hsv_img, None)
```
