### Tạo Tensor với `tf.Variable()`

sự khác biệt giữa tf.variable và tf.constant là tensor được tạo bởi tf.constant là bất biến không thay đổi, còn được tạo bởi tf.variable có thể thay đổi

```python
changeable_tensor = tf.Variable([10,7])
unchangeable_tensor = tf.constant([10,7])
changeable_tensor, unchangeable_tensor
```

để thay đổi phần tử của tensor tf.Variable cần phương thức assign

```python
changeable_tensor[0].assign(7)

changeable_tensor
```

việc lựa chọn tf.constant hay tf.Variable tùy thuộc vào yêu cầu nhưng đa số trường hợp TensorFlow sẽ tự động chọn giúp chúng ta

### Tạo các random tensor

```python
# tao 2 random tensor
random_1 = tf.random.Generator.from_seed(42) # thiet lap seed cho kha nang tai tao
random_1 = random_1.normal(shape=(3,2)) #tao tensor tu phan phoi chuan
random_2 = tf.random.Generator.from_seed(42)
random_2 = random_2.normal(shape=(3,2))

random_1, random_2, random_1 == random_2
```
